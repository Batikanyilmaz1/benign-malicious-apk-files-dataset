```python
import hashlib

def getHash(file_path):
  sha256 = hashlib.sha256()
  with open(file_path, 'rb') as f:
    for block in iter(lambda: f.read(65536), b''):
      sha256.update(block)
  return sha256.hexdigest()
```

```python
# Define the permissions you want as indexes
permissions = ['android.permission.ACCESS_BACKGROUND_LOCATION', 'android.permission.ACCESS_ADSERVICES_AD_ID', 'android.permission.ACCESS_NETWORK_STATE']

# Paths to APK files
benign_apk_files = [
    "./drive/MyDrive/benign_apks/MP3_Downloader.apk",
    "./drive/MyDrive/benign_apks/NYT_Games.apk",
    "./drive/MyDrive/benign_apks/Pixel_Camera.apk",
    "./drive/MyDrive/benign_apks/Planta.apk",
    "./drive/MyDrive/benign_apks/Premier_Agent.apk",
    "./drive/MyDrive/benign_apks/Salesforce.apk",
    "./drive/MyDrive/benign_apks/Signal.apk",
    "./drive/MyDrive/benign_apks/Stash.apk",
    "./drive/MyDrive/benign_apks/Tripadvisor.apk",
    "./drive/MyDrive/benign_apks/ZohoMail.apk"
]

malicious_apk_files = [
    "./drive/MyDrive/malware_apks/0f678cac9dadf0503670a3da1c57d5738bfbd2947d6bd40a8a557df3ec3caf23.apk",
    "./drive/MyDrive/malware_apks/41ad44002ea3545381d9a9142325626da071c7e0b2aa83ae94424a11563366da.apk",
    "./drive/MyDrive/malware_apks/55b9aa2a41f6fd46aa285d8a24f9dab0bdea729797199aa28648d89f036cff6b.apk",
    "./drive/MyDrive/malware_apks/5dfc459a0a89d8e48d4d6fc9cb5f572df9223b77508735726eeb5a1c4a5e96de.apk",
    "./drive/MyDrive/malware_apks/6c60af5424a82c7eb06d5e6db79c7b173d6a2cbcd31cf6f715112b34fbd8a22c.apk",
    "./drive/MyDrive/malware_apks/72d0ff2667eea0fc7173c1458c491bc5ec497c7eeb9ddc59ac5456223eea0634.apk",
    "./drive/MyDrive/malware_apks/798fafd34288fe8439dfd609d2599f234c6cf63b047f032d872d228f951c87ae.apk",
    "./drive/MyDrive/malware_apks/9201c036fe92744a9bf622e21af61d8be4e744ae3cb4b433705774ad6a0943f6.apk",
    "./drive/MyDrive/malware_apks/c6b4d58f604325a80db90a61da2cd0ce92d9be5c1c08c66b12403f04fd3b1676.apk",
    "./drive/MyDrive/malware_apks/d16dfc3f2a42feaa5598a589565684f351c589bee7a9393bc4102dcc0bc8a22d.apk"
]

# Create a dictionary to store permissions for each APK
permissions_dict = {}

# Iterate over benign APK files and extract permissions
for apk_file in benign_apk_files:
    apk = APK(apk_file)
    file_hash = getHash(apk_file)
    apk_permissions = apk.get_permissions()
    permissions_dict[file_hash] = pd.Series((1 if perm in apk_permissions else 0 for perm in permissions), index=permissions)

# Iterate over malicious APK files and extract permissions
for apk_file in malicious_apk_files:
    apk = APK(apk_file)
    file_hash = apk_file.split("/")[-1].split(".")[0]  # Extracting hash from filename
    apk_permissions = apk.get_permissions()
    permissions_dict[file_hash] = pd.Series((1 if perm in apk_permissions else 0 for perm in permissions), index=permissions)

# Create DataFrame from the dictionary
dataframe = pd.DataFrame(permissions_dict)

# Fill NaN values with 0
dataframe = dataframe.fillna(0)

# Display the DataFrame
dataframe
```
